```python
stream_df = pd.read_csv('dm_spotify.csv', 
                        usecols=['released_year', 'streams', 'streamsM', 'track_name', 'artist(s)_name']).head(250)
#streams_df =stream_df.head(400)
```

```python
# Using scatter hvplot in order to show the streaming data
stream_df.hvplot.scatter(
    x='released_year',
    y='streamsM',
    title='Example Stream Plot',
).opts(yformatter='%.0f', xformatter='%.0f')
```

# Linear Regression

```python
# Reshaping  the released column for 'X' variable
X = stream_df['released_year'].values.reshape(-1, 1)

X[:100]
```

```python
# Showing the shape of 'X'
X.shape
```

```python
# Assigning the the average of stream numbers into the variable 'y'
y= stream_df['streamsM']
```

```python
# Adding Algorithim to the model variable
model = LinearRegression()
```

```python
# Fitting X, Y variables for Algorithim
model.fit(X, y)
```

```python
# Print model slope from coefficient
print(f"Model's slope: {model.coef_}")
```

```python
# Adding the intercept and coefficient multiplied by 'X' for 'y'
print(f"Model's formula: y = {model.intercept_} + {model.coef_[0]}X")
```

```python
# Adding the intercept and coefficient multiplied by number of years
print(f"Model's formula: y = {model.intercept_} + {model.coef_[0]} * 10")
```

```python
# Calculating the predicted value
y_7 = model.intercept_ + model.coef_[0] * 5
```

```python
print(f"Predicted streams for aperson with 10 years of streaming music: {y_7:.2f}")
```

```python
# Printing predicted values from algorithim
predicted_y_values = model.predict(X)
print("%d", predicted_y_values)
```

```python
# Adding the predicted values to the dataframe
stream_predict_df = stream_df.copy()

stream_predict_df['stream predictions'] = predicted_y_values

stream_predict_df.head()
```

```python
#Linear Regression Algorithim line plot
stream_plot = stream_predict_df.hvplot.line(

    x='stream predictions',
    y='released_year',
    color='purple'
).opts(yformatter='%.0f', xformatter='%.0f')

stream_plot
```

```python
# Combining the line and scatter plot
stream_plot = stream_predict_df.hvplot.line(

    x='stream predictions',
    y='released_year',
    color='purple'
).opts(yformatter='%.0f', xformatter='%.0f') * stream_predict_df.hvplot.scatter(

    x='stream predictions',
    y='released_year',
    color='purple'
).opts(yformatter='%.0f', xformatter='%.0f')

stream_plot
```

```python
# KDE Plot for regression algorithim
stream_plot = stream_predict_df.hvplot.kde(

    #x='stream predictions',
    y='released_year',
    color='purple',
    subplots=True
).opts(yformatter='%.0f', xformatter='%.0f')

stream_plot
```
